```python
# Simple line graph
x_val = [0, 1, 2]
y_val = [3, 4, 5]
plt.plot(x_val, y_val)
plt.show()
```

```python
# a simple graph without x values; it assumes xticks as 0, 1, 2, .....
y_val = [3, 5, 7]
plt.plot(y_val)
plt.show()
```

```python
# but y axis values are must; if we don't provide, it automatically assumes given values as y values and for x-axis it goes with default 0, 1, 2, ....
x_val = [3, 5, 8]
plt.plot(x_val,)
plt.savefig('Simple Graph')
```

```python
# if we use plt.show right after plt.plot() it either gives separate plot or plots together in one plot
x_val = [0, 1, 2]
y_val = [3, 5, 7]
plt.plot(x_val, y_val)
plt.show()
x_val1 = [0, 1, 2]
y_val1 = [5, 9, 11]
plt.plot(x_val1, y_val1)
plt.show()
```

```python
# here we use plt.show at the end and thus get one single plot with two curves
x_val = [0, 1, 2]
y_val = [3, 5, 7]
plt.plot(x_val, y_val)
x_val1 = [0, 1, 2]
y_val1 = [5, 9, 11]
plt.plot(x_val1, y_val1)
plt.show()
```

```python
x_val = [-2, -1, 0, 1, 2]
y_val = [4, 1, 0, 1, 4]
plt.plot(x_val, y_val)
plt.show()
```

```python
# to make the above parabola graph smoother, use more points
x = np.arange(-2, 2, step=0.1)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
# Use xticks to show the x axis actual values and to display the units 'rotation' attribute can be used
x = np.arange(-2, 2, step=0.15)
y = x**2
plt.plot(x, y)
plt.xticks(x, rotation=90)
plt.show()
```

```python
# adding Title, and Labels to x & y axis
x = np.arange(0, 5)
y = x**2
plt.plot(x, y)
plt.xlabel('x_values')
plt.ylabel('x**2')
plt.title('Simple Graph')
plt.show()
```

```python
# to change the color of the curve, use color attribute in plt.plot
x = np.arange(0, 5)
y = x**2
plt.plot(x, y, color='green')
plt.xlabel('x_values')
plt.ylabel('x**2')
plt.title('Simple Graph')
plt.show()
```

```python
# to make the curve as dotted or star'ed use linestyle attribute in plt.plot
x = np.arange(0, 5, step=0.2)
y = x**2
plt.plot(x, y, '*')
plt.xlabel('x_values')
plt.ylabel('x**2')
plt.title('Multiline Graph')
x1 = np.arange(0, 5, step=0.2)
y1 = x1**3
plt.plot(x1, y1, 'r--')
x2 = x
y2 = x2**2.5
plt.plot(x2, y2, 'go')
plt.show()
```

```python
# to zoom in a particular portion of the graph; here in this case only right portion of the graph
x = np.arange(-2, 2, step=0.1)
y = x**2
plt.plot(x, y, '*')
plt.xlim(left=0, right=2)
plt.show()
```

```python
# to zoom in a particular portion of the graph; here in this case only left portion of the graphy
x = np.arange(-2, 2, step=0.1)
y = x**2
plt.plot(x, y, '*')
plt.xlim(left=-2, right=0)
plt.show()
```

```python
# to zoom in a particular portion of the graph; some random middle portion
x = np.arange(-2, 2, step=0.1)
y = x**2
plt.plot(x, y, '*')
plt.xlim(left=-1, right=1)
plt.show()
```

```python
# adding legends to graphs
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
plt.plot(x3, y3, label="Squares")
plt.plot(x3, y4, label="Cubes")
plt.legend()
plt.show()
```

```python
# another way to add Legend
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
plt.plot(x3, y3)
plt.plot(x3, y4)
plt.legend(['Squares', 'Cubes'])
plt.show()
```

```python
# adding Legend at the bottom corner
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
plt.plot(x3, y3)
plt.plot(x3, y4)
plt.legend(['Squares', 'Cubes'], loc='lower right')
plt.show()
```

```python
# adding Legend at the right middle
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
plt.plot(x3, y3)
plt.plot(x3, y4)
plt.legend(['Squares', 'Cubes'], loc='center right')
plt.show()
```

```python
# adding Legend at the bottom
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
plt.plot(x3, y3)
plt.plot(x3, y4)
plt.legend(['Squares', 'Cubes'], loc='lower center')
plt.show()
```

```python
# Legend at the bottom overlapping the graph
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
y5 = x3**4
plt.plot(x3, y3)
plt.plot(x3, y4)
plt.plot(x3, y5)
plt.legend(['Squares', 'Cubes', 'Quadrapules'], loc='lower center')
plt.show()
```

```python
# Resolving the overapping legend at the bottom
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
y5 = x3**4
plt.plot(x3, y3)
plt.plot(x3, y4)
plt.plot(x3, y5)
plt.legend(['Squares', 'Cubes', 'Quadrapules'], loc='lower center', ncol=3)
plt.show()
```

```python
# frameon attribute shows/hides the outer line/border of the legend
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
y5 = x3**4
plt.plot(x3, y3)
plt.plot(x3, y4)
plt.plot(x3, y5)
plt.legend(['Squares', 'Cubes', 'Quadrapules'], loc='lower center', ncol=3, frameon=False)
plt.show()
```

```python
# to add Grid to the graph
x3 = np.arange(-5, 5, step=0.1)
y3 = x3**2
y4 = x3**3
y5 = x3**4
plt.plot(x3, y3)
plt.plot(x3, y4)
plt.plot(x3, y5)
plt.legend(['Squares', 'Cubes', 'Quadrapules'], loc='lower center', ncol=3, frameon=False)
plt.grid()
plt.show()
```

```python
# to plot a bar chart
plt.bar(['Apple', 'Orange', 'Mango'], [200, 110, 125])
plt.show()
```

```python
# to change the width of the bar
plt.bar(['Apple', 'Orange', 'Mango'], [200, 110, 125], width=0.5)
plt.show()
```

```python
# Add title, labels to the chart
plt.bar(['Apple', 'Orange', 'Mango'], [200, 110, 125], width=0.5)
plt.title("Fruits Prices")
plt.xlabel("Fruits")
plt.ylabel("Price")
plt.show()
```

```python
# to plot two bars for comparison
plt.bar([1, 2, 3], [10, 20, 15], width=0.2, label='Market1')
plt.bar(np.array([1, 2, 3])+0.2, [12, 18, 13], label='Market2', width=0.2)
plt.legend()
plt.show()
```

```python
# to plot two bars for comparison
plt.bar([1, 2, 3], [10, 20, 15], width=0.2, label='Market1', bottom=None)
plt.bar(np.array([1, 2, 3])+0.2, [12, 18, 13], label='Market2', width=0.2, bottom=None)
plt.legend()
plt.show()
```

```python
# to plot a histogram. Histo chart gives the frequency bars of each value
values = [1, 2, 3, 4, 3, 4, 1, 3, 2, 4, 2, 4]
plt.hist(values)
plt.show()
```

```python
# to categorize the histogram values, use bins
values = [1, 2, 3, 4, 3, 4, 1, 3, 2, 4, 2, 4]
plt.hist(values, bins=2)
plt.show()
```

```python
# to categorize the histogram values, use bins
values = [1, 2, 3, 4, 3, 4, 1, 3, 2, 4, 2, 4]
plt.hist(values, bins=[1,3,4])
plt.show()
```

```python
# to categorize the histogram values, use bins
values = [1, 2, 3, 4, 3, 4, 1, 3, 2, 4, 2, 4]
plt.hist(values, bins=[1,2,3,4])
plt.show()
```

```python
# to plot a scatter plot, use plt.scatter()
xval = np.arange(0, 20)
yval = np.linspace(1, 20, num=20) + np.random.rand(20)*10
plt.scatter(xval, yval)
plt.show()
```

```python
# to plot a scatter plot, use plt.scatter()
xval = np.arange(0, 20)
yval = np.linspace(1, 20, num=20) + np.random.rand(20)*10
yval2 = np.linspace(1, 20, num=20) + np.random.rand(20)*20
plt.scatter(xval, yval, label="Store1")
plt.scatter(xval, yval2, label="Store2")
plt.legend()
plt.show()
```
