## Some minor data pre-processing

```python
# fill NAs with 0s
df = df.fillna(1)
```

```python
df.isna().sum()
```

```python
df["CO2 Per Capita (metric tons)"]
```

```python
#  Make DataFrame Pipeline Interactive
idf = df.interactive()
```

```python
# Define Panel Widgets
year_slider = pn.widgets.IntSlider(name='Year Slider', start=1960, end=2011, step=5, value=1970)
year_slider
```

```python
df['CO2 Per Capita (metric tons)'].mean()
```

```python
# Radio buttons for CO2 measures
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name = 'Y axis',
    options = ['CO2 Per Capita (metric tons)'],
    button_type = 'success'
)
```

```python
yaxis_co2
```

```python
continents = ['Latin America & Caribbean', 'Europe & Central Asia', 'South Asia',
       'Sub-Saharan Africa', 'Middle East & North Africa',
       'East Asia & Pacific', 'North America']
```

```python
continents
```

```python
co2_pipeline = (
    idf[
        (idf.Year <= year_slider)
    ]
    .groupby(['Country Name', 'Year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')
    .reset_index(drop=True)
)
```

```python
co2_pipeline
```

```python
co2_plot = co2_pipeline.hvplot(x='Year', by='Country Name', y=yaxis_co2, line_width=2, title="CO2 Emission by Contient")
```

```python
co2_plot
```

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size=10, sizing_mode='stretch_width')
```

```python
co2_table
```

```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.Year == year_slider)
    ]
    .groupby(['Country Name', 'Year', 'CO2 (kt)'])['CO2 Per Capita (metric tons)'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')
    .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```
