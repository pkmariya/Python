```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import urllib
import requests
import numpy as np
import matplotlib.pyplot as plt
```

```python
words = 'access guest guest apartment area bathroom guest bed bed bed bed bedroom block coffee coffee entrance entry free francisco garden kebi guest home house kettle kitchen living microwave neighbourhood new park parking place privacy private queen room san separate share space street suite time welcome '
```

```python
my_words = 'Mariya Preeths San francisco Kul Phil Tom Mariya Preeths Preeths Kulo Tiruppur Coimbatore Chennai Chennai Chennai Chennai Chennai Austin Florida San francisco Coimbatore'
```

```python
mask = np.array(Image.open(requests.get('http://www.clker.com/cliparts/O/i/x/Y/q/P/yellow-house-hi.png', stream=True).raw))
```

```python
# Function to take the words and mask to create the wordcloud
def create_wordcloud(words, mask):
    word_cloud = WordCloud(background_color='white', width=512, height=512, max_words=50, mask=mask, stopwords=set(STOPWORDS))
    plt.figure(figsize=(10, 8), facecolor='k', edgecolor='blue')
    plt.imshow(word_cloud.generate(words))
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```

```python
create_wordcloud(words, mask)
```

```python
create_wordcloud(my_words, mask)
```
