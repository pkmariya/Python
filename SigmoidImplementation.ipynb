# ML Regression Class1 - Sigmoid implementation variations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1 + (np.exp(-x)))
```

```python
x = np.linspace(-10, 10, 20) # 20 equidistant points between -10 and 10
z = sigmoid(x)

plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
plt.grid()
plt.show()
```

```python
def sigmoid1(x):
  return [1 / (1 + np.exp(-ele)) for ele in x]
```

```python
x = np.linspace(-10, 10, 20) # 20 equidistant points between -10 and 10
z = sigmoid1(x)

plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
plt.grid()
plt.show()
```

```python
def sigmoid2(x):
  min_limit = 709
  return 1 / (1 + np.exp(-(np.clip(x, -min_limit, None))))
```

```python
x = np.linspace(-10, 10, 20) # 20 equidistant points between -10 and 10
z = sigmoid2(x)

plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
plt.grid()
plt.show()
```
